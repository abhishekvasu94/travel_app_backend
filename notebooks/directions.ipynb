```python
distance_between_cities = []

for pair in tqdm(pairs):

    city_1 = pair[0]
    city_2 = pair[1]

    # try:
    #     transit_distance, transit_duration = get_dist_and_time(city_1, city_2, "transit", gmaps)
    # except IndexError:
    #     transit_distance, transit_duration = 0, 0

    # try:
    #     driving_distance, driving_duration = get_dist_and_time(city_1, city_2, "driving", gmaps)
    # except IndexError:
    #     transit_distance, transit_duration = 0, 0

    try:
        transit_leg = get_dist_and_time(city_1, city_2, "transit", gmaps)
    except IndexError:
        transit_leg = []

    try:
        driving_leg = get_dist_and_time(city_1, city_2, "driving", gmaps)
    except IndexError:
        driving_leg = []

    distance_dict = {
        "city_1": city_1, 
        "city_2": city_2,
        "transit_leg": transit_leg,
        "driving_leg": driving_leg
    }

    distance_between_cities.append(distance_dict)
```

```python
with open("../data/distance_between_cities.json", "w") as f:
    json.dump(distance_between_cities, f)
    f.close()
```
